```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
age_gender_datasets =pd.read_csv("/kaggle/input/age-gender-and-ethnicity-face-data-csv/age_gender.csv")
age_gender_datasets.info()
```

```python
# dataset filter
age_data_select = age_gender_datasets.dropna(subset ='age')
age_data_select = age_data_select[age_data_select['age']>0]

sample_dataset_for_sampling, _ = train_test_split(age_data_select,
                                              test_size =.2,
                                              stratify = age_data_select['gender'],
                                              random_state =42)
print(len(sample_dataset_for_sampling))
                                               
```

```python
sns.set(style ='whitegrid')
plt.figure(figsize = (12,6))
sns.histplot(data = sample_dataset_for_sampling,
             
             x = 'age',
             hue = 'gender',
             bins = 20,
             kde = True,
             palette = 'Set2',
             stat = 'density',
             common_norm = False
            
            )
plt.title("Stratify Data Sampling")
plt.xlabel("Age data")
plt.ylabel("Gender Data")
plt.legend(title ="Gender",
          labels =["Male(1)","Female(0)"])
plt.tight_layout()
plt.show()
```

```python
# credit card dataset
credit_card_dataset = pd.read_csv('/kaggle/input/credit-data/credit_data.csv')
credit_card_dataset.info()
```

```python
credit_card_dataset = credit_card_dataset.dropna(subset = 'age')
nth_interval_value = 8
starting = np.random.randint(0,nth_interval_value )
print(starting)
systematic_sampling = credit_card_dataset.iloc[starting::nth_interval_value]
print("systematic_sampling\n",systematic_sampling)
```

```python
sns.set(style ='whitegrid')
plt.figure(figsize = (12,6))
sns.histplot(systematic_sampling['age'], #selected data visualization
             bins = 30,
             kde = True,
             color = '#886544'
            
            )
plt.title("Systematic Data Sampling")
plt.xlabel("Age data")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
